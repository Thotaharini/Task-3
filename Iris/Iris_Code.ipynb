```python


import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")

print("\n==============================")
print("CREATING DATASETS (NO FILE ERRORS)")
print("==============================")

netflix = pd.DataFrame({
    "type": ["Movie", "TV Show", "Movie", "TV Show", "Movie"],
    "release_year": [2016, 2018, 2020, 2019, 2021],
    "rating": ["PG", "TV-MA", "R", "TV-14", "PG-13"]
})

print("Netflix dataset ready")

iris = pd.DataFrame({
    "sepal_length": [5.1, 4.9, 6.2, 5.9, 6.5],
    "sepal_width": [3.5, 3.0, 3.4, 3.0, 3.2],
    "petal_length": [1.4, 1.4, 5.4, 5.1, 5.8],
    "petal_width": [0.2, 0.2, 2.3, 1.8, 2.2],
    "species": ["setosa", "setosa", "virginica", "virginica", "virginica"]
})

print("Iris dataset ready")


print("\n==============================")
print("NETFLIX EDA")
print("==============================")

plt.figure()
plt.hist(netflix["release_year"], bins=5)
plt.title("Distribution of Release Year")
plt.xlabel("Release Year")
plt.ylabel("Count")
plt.show()

print("INSIGHT: Most Netflix content is released after 2016.")

plt.figure()
sns.countplot(x="type", data=netflix)
plt.title("Movies vs TV Shows")
plt.show()

print("INSIGHT: Netflix platform contains more Movies than TV Shows.")

print("\n==============================")
print("IRIS DATASET EDA")
print("==============================")

iris.drop("species", axis=1).hist(figsize=(8,6))
plt.suptitle("Distribution of Numerical Features")
plt.show()

print("INSIGHT: Petal measurements vary more than sepal measurements.")

plt.figure(figsize=(8,5))
sns.boxplot(data=iris.drop("species", axis=1))
plt.xticks(rotation=45)
plt.title("Box Plot  Outlier Detection")
plt.show()

print("INSIGHT: Mild outliers observed in petal features.")

plt.figure(figsize=(6,5))
sns.heatmap(
    iris.drop("species", axis=1).corr(),
    annot=True,
    cmap="coolwarm"
)
plt.title("Correlation Heatmap")
plt.show()

print("INSIGHT: Strong correlation between petal length and petal width.")

print("\n==============================")
print("FEATURE IMPORTANCE")
print("==============================")
print(" Petal length and petal width are most important for prediction.")
print(" High correlation indicates multicollinearity.")

print("\n==============================")
print("FINAL SUMMARY")
print("==============================")
print("""
 EDA helps understand data patterns and relationships.
 Visualization makes insights easier to interpret.
 Netflix content shows recent growth trend.
 Iris petal features are strong predictors.
 Correlation analysis helps identify multicollinearity.
""")
```
